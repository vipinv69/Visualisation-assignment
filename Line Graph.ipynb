```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv('/content/prices.csv')
```

```python
def lin_diag(gasp):



  plt.figure(figsize=(8, 5)) #adjusting size of figure

  plt.title('Gas Prices in USD', fontdict = {'fontweight':'bold', 'fontsize':16}) #Giving Title to the Graph
  plt.plot(gasp.Year, gasp.France, label='France') 
  plt.plot(gasp.Year, gasp.Canada, label='Canada') #plotting line graph of countries
  plt.plot(gasp.Year, gasp.UK, label='United Kingdom') 

  plt.xticks(prices.Year[::10])    #setting x-axis

  plt.xlabel('Year') 
  plt.ylabel('US Dollars')
  plt.legend()
  plt.show()

lin_diag(prices)
```
