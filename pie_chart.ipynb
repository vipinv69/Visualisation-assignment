```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
football = pd.read_csv('/content/fifa21.csv')
```

```python
plt.style.use('ggplot') #style usage

beginner = football.loc[football.OVA <= 60].count()[0]
subjunior = football.loc[(football.OVA > 60) & (football.OVA <= 65)].count()[0]  #dividing categories of overall range
junior = football.loc[(football.OVA > 65) & (football.OVA <= 70)].count()[0]
semipro = football.loc[(football.OVA > 70) & (football.OVA <= 75)].count()[0]
pro = football.loc[(football.OVA > 75) & (football.OVA <= 80)].count()[0]
elite = football.loc[(football.OVA > 80) & (football.OVA <= 100)].count()[0]


overall = [beginner, subjunior, junior, semipro, pro, elite]
labels = ['Under 60', 'Between 61-65', 'Between 66-70', 'Between 71-75', 'Between 76-80', 'Over 80']
explode = (0, 0, 0, 0, 0.1, 0.4)  #explode
plt.figure(figsize=(15,10)) #increse size
plt.pie(overall, labels = labels, autopct = '%.2f%%', pctdistance = 0.65, explode = explode, wedgeprops = {'edgecolor':'black'}) #plotting pie chart
plt.title('Distribution of player ratings in FIFA 2021', fontdict = {'fontweight':'bold', 'fontsize':16}) #create title
plt.legend()
plt.show()

```
